```python
import csv
import numpy as np
import matplotlib.pyplot as plt

with open("470_8.csv") as f:
    reader = csv.reader(f)
#    next(reader) # skip header
    data = [r for r in reader]

print(data[:10])
```

```python
frame_rate = 37
x_data = []
y_data = []
for i in range(len(data)):
    x_data.append(i/frame_rate)
    y_data.append(float(data[i][0]))
    
plt.plot(x_data,y_data,marker='o',color='b')
#plt.xlim(0,60)
plt.show()
```
